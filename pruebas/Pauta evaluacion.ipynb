```python
# Respuesta

def ejercicio3(lista,letra):
    #Transformar letra en lower case
    letra=letra.lower()
    #detectamos letra mala y avisamos
    if len(letra) != 1:
        print('Letra con largo mayor a uno')
        return
    #detectamos lista mala y avisamos
    for i in lista:
        #if not isinstance(i, str):
        if type(i)!=str:
            print('Elemento ', i, ' de la lista no es string')
            return

    l = []
    for i in lista:
        if i[0] == letra:
            l.append(i)
    
    return l

print(ejercicio3(nombres,letra))
```
